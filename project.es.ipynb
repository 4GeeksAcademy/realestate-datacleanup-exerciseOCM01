```python
ds['price'] = pd.to_numeric(ds['price'], errors='coerce')
ds['surface'] = pd.to_numeric(ds['surface'], errors='coerce')

ds['pps'] = ds['price'] / ds['surface']

promedios_pps = ds.groupby('level5')['pps'].mean()

# Extraer los valores para Valdemorillo y Galapagar
valdemorillo_pps = promedios_pps.get('Valdemorillo')
galapagar_pps = promedios_pps.get('Galapagar')

# Imprimir resultados
print(f"Precio promedio por m² en Valdemorillo: {valdemorillo_pps:.2f} €")
print(f"Precio promedio por m² en Galapagar:    {galapagar_pps:.2f} €")
```

#### Ejercicio 18. Normaliza la variable de precios para cada población y traza los 4 histogramas en el mismo gráfico (debes usar el subconjunto obtenido en la pregunta 14) (★★★)

Para el método de normalización, puedes usar el que consideres adecuado, no hay una única respuesta correcta para esta pregunta. Imprime el gráfico y escribe en la celda de Markdown un breve análisis sobre el gráfico.

Pista: Puedes ayudarte revisando la demostración multihist de Matplotlib.

```python
df_sur['price_norm'] = (
    df_sur
    .groupby('level5')['price']
    .transform(lambda x: (x - x.mean()) / x.std())
)

# Visualización
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
bins = 30

for municipio in df_sur['level5'].unique():
    datos = df_sur.loc[df_sur['level5'] == municipio, 'price_norm'].dropna()
    plt.hist(
        datos,
        bins=bins,
        density=True,
        alpha=0.5,
        label=municipio
    )

# Línea vertical en z=0 (media local)
plt.axvline(0, color='black', linestyle='--', linewidth=1)

plt.title('Histogramas de precios normalizados\npor municipio (Cinturón Sur de Madrid)')
plt.xlabel('Precio normalizado (z-score)')
plt.ylabel('Densidad')
plt.legend(title='Municipio')
plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```

#### Ejercicio 21. Realiza un trazado de las coordenadas (columnas latitud y longitud) del cinturón sur de Madrid por color de cada población (debes usar el subconjunto obtenido del Ejercicio 14) (★★★★)

Ejecuta la siguiente celda y luego comienza a codear en la siguiente. Debes implementar un código simple que transforme las columnas de coordenadas en un diccionario de Python (agrega más información si es necesario) y agrégala al mapa.

```python
from ipyleaflet import Map, basemaps

# Mapa centrado en Madrid
map = Map(center = (40.35, -3.8), zoom = 11.5, min_zoom = 1, max_zoom = 20, 
    basemap=basemaps.OpenStreetMap.Mapnik)
map
```

```python
from ipyleaflet import Map, MarkerCluster, CircleMarker, basemaps, LayersControl

ds.columns = ds.columns.str.strip()

for col in ['latitude', 'longitude']:
    ds[col] = (
        ds[col]
        .astype(str)
        .str.replace(',', '.', regex=False)
        .astype(float)
    )

cinturon_sur = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]
df_sur = ds[ds['level5'].isin(cinturon_sur)].copy()

colors = {
    'Fuenlabrada': 'green',
    'Leganés':     'blue',
    'Getafe':      'red',
    'Alcorcón':    'orange'
}

markers = []
for muni, group in df_sur.groupby('level5'):
    for lat, lon in zip(group['latitude'], group['longitude']):
        markers.append(
            CircleMarker(
                location=(lat, lon),
                radius=4,
                color=colors[muni],
                fill_color=colors[muni],
                fill_opacity=0.6,
                tooltip=muni
            )
        )

m = Map(
    center=(40.35, -3.8),
    zoom=11.5,
    min_zoom=1,
    max_zoom=20,
    basemap=basemaps.OpenStreetMap.Mapnik
)

cluster = MarkerCluster(markers=markers)
m.add_layer(cluster)

from ipyleaflet import LayersControl
m.add_control(LayersControl())

m
```
